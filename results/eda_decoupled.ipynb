Charging flashing time = Time from voltage minimum (start of charging) to when oxygen drop becomes stable.

Discharging flashing time = Time from voltage maximum (start of discharging) to when oxygen increasing upto the initial value.

# Flashing rate 2 ml

```python
from scipy.signal import argrelextrema

dfVol = df_plot_clean2['Value '].values
dfOX = df_plot_clean2['Sensor 1 - OX (μmol/L)'].values
time = df_plot_clean2['Time_HMS_dt'].values

sampling_interval_sec = 1
window_plateau_min = 2
window_plateau = int(window_plateau_min * 60 / sampling_interval_sec)
initial_value = 250
min_vol_threshold = -1.29
max_vol_threshold = 1.8
window_stable = 20
charging_times = []
discharging_times = []
```

```python
min_vol_indices = np.where(dfVol <= min_vol_threshold)[0]

min_vol_indices = min_vol_indices[np.insert(np.diff(min_vol_indices) > 60, 0, True)]
for idx in min_vol_indices:
    for j in range(idx, min(len(dfOX)-window_stable, idx+3000)):
        grad = np.abs(np.gradient(dfOX[j:j+window_stable]))
        if grad.mean() < 2.0:
            t0 = pd.to_datetime(time[idx])
            t1 = pd.to_datetime(time[j])
            duration = (t1 - t0).total_seconds() / 60
            charging_times.append({'start': t0, 'end': t1, 'duration_min': duration})
            break
```

```python
max_vol_indices = np.where(dfVol >= max_vol_threshold)[0]

for idx in max_vol_indices:
    for j in range(idx+1, len(dfOX)-window_plateau):
        if np.all(dfOX[j:j+window_plateau] >= initial_value):
            t0 = pd.to_datetime(time[idx])
            t1 = pd.to_datetime(time[j])
            duration = (t1 - t0).total_seconds() / 60
            discharging_times.append({'start': t0, 'end': t1, 'duration_min': duration})
            break
```

```python
charging_df = pd.DataFrame(charging_times)
discharging_df = pd.DataFrame(discharging_times)
```

```python
charging_df = charging_df[
    (charging_df['duration_min'] >= 1) & (charging_df['duration_min'] <= 6)
].reset_index(drop=True)

discharging_df = discharging_df[discharging_df['duration_min'] >= 2].reset_index(drop=True)
```

```python
print("Charging loop (first 5):\n", charging_df.head())
print("Discharging loop (first 5):\n", discharging_df.head())
```

```python
charging_df['duration_min'].mean()
```

```python
discharging_df['duration_min'].mean()
```

```python
all_durations = pd.concat([
    charging_df['duration_min'],
    discharging_df['duration_min']
], ignore_index=True)

mean_all = all_durations.mean()

print(f"\nAverage time with Flow rate 2 ml (All Loops): {mean_all:.2f} minutes")
```
